```python
print(data.info())
```

```python
stats=data['Population Vaccinated'].describe()
stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Population Vaccinated'])
plt.title('Box plot of population vaccinated')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Population Vaccinated'], kde=True)
plt.title('Histogram of population vaccinated')
plt.xlabel('Population Vaccinated')
plt.show()
```

```python
Q1 = data['Population Vaccinated'].quantile(0.25)
Q3 = data['Population Vaccinated'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

potential_outliers = data[(data['Population Vaccinated'] < lower_bound) | (data['Population Vaccinated'] > upper_bound)]
print(potential_outliers)
```
