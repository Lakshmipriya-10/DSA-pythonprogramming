**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np

data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)

print(df)
````

**loading diabetes csv file**

````python
df1 = pd.read_csv('diabetes_model .csv')
df1
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

**univariate analysis using histogram**

````python
columns = df1.columns

for col in columns:
    plt.figure(figsize=(8, 4))
    plt.hist(df1[col], bins=20)
    plt.title(f'Histogram of {col}')
    plt.show()
````

**univariate analysis using boxplots**

````python
import seaborn as sns
for col in columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df1[col])
    plt.title(f'Box Plot of {col}')
    plt.show()
````

**Bivariate Analysis:scatterplot between pairs of variables**

````python
plt.figure(figsize=(8,6))
plt.scatter(df1['Pregnancies'],df1['Age'])
plt.show()
````

````python
plt.figure(figsize=(8, 6))
plt.scatter(df1['Age'], df1['BloodPressure'], color='purple')
plt.show()
````

````python
plt.figure(figsize=(8,6))
plt.scatter(df1['Glucose'], df1['BMI'],color = 'green')
plt.show()
````

````python
plt.figure(figsize=(8,6))
plt.scatter(df1['SkinThickness'], df1['Insulin'],color = 'red')
plt.show()
````

**correlation matrix**

````python
plt.figure(figsize=(8, 6))
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
plt.show()
````
